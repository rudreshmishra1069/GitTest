```python
#route: It is the concatenation of 'from_station_id' and 'to_station_id'

# Hence, each route is a unique id Identifying the route of a particular trip
# Now to find the most common routes of each year, we will find the occurence count of each unique route

unique_route_occurence_count_2018 = dataset_2018.groupby(['route'])['route'].count().reset_index(name='count')
unique_route_occurence_count_2017 = dataset_2017.groupby(['route'])['route'].count().reset_index(name='count')
```

```python
print(unique_route_occurence_count_2017)
```

```python
print(unique_route_occurence_count_2018)
```

```python
# Now to find the most used routes, we'll sort the 'unique_route_occurence_count_2017' and 'unique_route_occurence_count_2018'
# by the 'count' coloumn. Then we'll take the top 1000 most frequently occuring routes from each year.

top_1000routes_2017 = unique_route_occurence_count_2017.sort_values(by = ['count'], ascending = False).head(1000)
top_1000routes_2018 = unique_route_occurence_count_2018.sort_values(by = ['count'], ascending = False).head(1000)
```

```python
print(top_1000routes_2017)
```

```python
print(top_1000routes_2018)
```

```python
# Now we'll create two lists of most common routes of 2017 and 2018

list_mostFreqRoutes_2017 = top_1000routes_2017['route'].to_list()
list_mostFreqRoutes_2018 = top_1000routes_2018['route'].to_list()
```
